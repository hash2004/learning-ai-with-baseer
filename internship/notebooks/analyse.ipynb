```python
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv("output-average-loudness_20241128_005658.csv")
data["Datetime"]=pd.to_datetime(data["Datetime"], errors="coerce")
plt.figure(figsize=(12, 6))
trend=data["Loudness (dB)"].rolling(window=100, min_periods=1).mean()
plt.plot(data["Datetime"], trend)
plt.title("Average Loudness Over Time")
plt.xlabel("Datetime")
plt.ylabel("Loudness (dB)")
plt.legend()
plt.tight_layout()
plt.axhline(-50, color="red", linestyle="--", label="Silence Threshold (-50 dB)")
plt.show()
```
